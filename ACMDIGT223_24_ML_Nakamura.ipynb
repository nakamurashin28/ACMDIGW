```python
df.drop(columns = ['Over18', 'EmployeeNumber', 'StandardHours', 'EmployeeCount'], inplace = True)
```

```python
#categorical encoding

df['Attrition'] = df['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)
df['OverTime'] = df['OverTime'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
# check valeus of attrition
df['Attrition'].value_counts()
```

```python
# check valeus of Overtime
df['OverTime'].value_counts()
```

```python
#create a histogram
df.hist(bins = 30, figsize = (20, 20), color = 'b')
```

```python
# create a filter of how many employees left vs stayed

emp_left = df[df['Attrition'] == 1]
emp_stay = df[df['Attrition'] == 0]
```

```python
print(f'Employees that left the MSFT ={len(emp_left)}')
print(f'Employees that stayed the MSFT ={len(emp_stay)}')
print(f'% Attrition {len(emp_left)/ len(df) * 100}%')
```
